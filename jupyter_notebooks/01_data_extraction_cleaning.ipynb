## 3. Conduct descriptive statistics on the cleaned dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the CSV file
df = pd.read_csv('../data/Online Retail.csv')

# Step 2: Display the first few rows and basic information about the dataset
print("First few rows of the dataset:")
print(df.head())
print("\nDataFrame Information:")
print(df.info())

# Step 3: Handle missing values
# Check for missing values before dropping
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)

# Drop rows with missing values
df.dropna(inplace=True)

# Step 4: Convert InvoiceDate to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Step 5: Create TotalTransactionValue column
df['TotalTransactionValue'] = df['Quantity'] * df['UnitPrice']

# Step 6: Remove negative quantities and prices
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Step 7: Save the cleaned dataset to a new CSV file
df.to_csv('../data/cleaned_online_retail.csv', index=False)

# Optional: Display descriptive statistics of the cleaned dataset
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset with specified dtype to avoid DtypeWarning
df = pd.read_csv('../data/cleaned_online_retail.csv', dtype={'InvoiceNo': str}, low_memory=False)

# Check if 'seaborn' style is available
print("Available styles:", plt.style.available)

# Set style for better-looking plots
plt.style.use('seaborn')

# Plot total sales over time
plt.figure(figsize=(12, 6))
df.groupby('InvoiceDate')['TotalTransactionValue'].sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.savefig('../figures/sales_over_time.png')  # Save the figure
plt.show()  # Display the plot

# Top 10 products by quantity sold
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
top_products.plot(kind='bar')
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.savefig('../figures/top_10_products.png')  # Save the figure
plt.show()  # Display the plot

# Customer distribution by country
plt.figure(figsize=(12, 6))
df['Country'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Customer Distribution by Country')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.savefig('../figures/customer_distribution.png')  # Save the figure
plt.show()  # Display the plot
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the cleaned dataset
df = pd.read_csv('../data/cleaned_online_retail.csv', dtype={'InvoiceNo': str}, low_memory=False)

# Ensure the figures directory exists
os.makedirs('../figures', exist_ok=True)

# 1. Monthly Sales Trend
plt.figure(figsize=(12, 6))
df['Month'] = pd.to_datetime(df['InvoiceDate']).dt.to_period('M')
monthly_sales = df.groupby('Month')['TotalTransactionValue'].sum()
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.savefig('../figures/monthly_sales_trend.png')
plt.close()

# 2. Top 10 Countries by Sales
top_countries = df.groupby('Country')['TotalTransactionValue'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar')
plt.title('Top 10 Countries by Sales')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../figures/top_10_countries.png')
plt.close()

# 3. Customer Purchase Frequency
purchase_frequency = df.groupby('CustomerID').size()
plt.figure(figsize=(12, 6))
sns.histplot(purchase_frequency, kde=True)
plt.title('Customer Purchase Frequency')
plt.xlabel('Number of Purchases')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.savefig('../figures/customer_purchase_frequency.png')
plt.close()

# 4. Product Price Distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x='UnitPrice', data=df)
plt.title('Product Price Distribution')
plt.xlabel('Unit Price')
plt.tight_layout()
plt.savefig('../figures/product_price_distribution.png')
plt.close()

# 5. Correlation Heatmap
numeric_columns = ['Quantity', 'UnitPrice', 'TotalTransactionValue']
correlation = df[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('../figures/correlation_heatmap.png')
plt.close()

print("Additional graphs have been generated and saved in the '../figures' directory.")
```
